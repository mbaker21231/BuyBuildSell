# Simple Models of Ownership, Enclosure, and Market Emergence

The idea here is to flush out some very simple models of how the three things mentioned above interrelate.

# Model I - Discrete choice - build? buy? 

Agents are born into the model not knowing their skill level at farming, which is ultimately determined by a draw from a distribution $f(\theta)$. 

Agents can, however, create a unit of an asset at some cost. The costs of creation has a common component, $c$, and an idiosyncratic component $e_c$ distributed according to $g(e)$.  

After the creation decision, agents observe productivities, and engage in trade. 

### Solution

Here is what I think happens in this model: agents with sufficiently high creation costs will not create an asset, but will instead wait for a purchase opportunity. So, the net returns to creating an asset are given by something like:

$$
E[\theta|\theta \geq p]P(\theta\geq p)+pP(\theta<p)-c_i
$$

The above anticipates that there is a threshold productivity level for which the agent keeps the asset and produces, which is in fact just $p$, the price of the asset. If it turns out that productivity is below price the agent just sells the asset and gets the price. 

Now, if the agent does not create a unit of the asset (because costs are too high), the agent might turn out to be a great producer, in which case the agent can jump back into the market and buy. Clearly, any agent who did not produce an asset in the first period has the chance to do this, and earn returns 

$$
P(\theta \geq p)\left(E[\theta|\theta \geq p]-p\right)
$$

Equilibrium depends upon three things $\theta_1,\theta_2,$ and $p$. This is a hard model to solve, but not impossible. Note that we can write the first equation as:

$$
\int_{p}^\infty \theta' f(\theta')d\theta'+pF(p)-c_i
$$

and the second equation as:

$$
\int_{p}^\infty \theta'f(\theta')d\theta'-(1-F(p))p
$$

By equating the first and second equation, we now have the condition that determines the cutoff cost above which agents will not produce an asset:
$$
\int_{p}^\infty \theta' f(\theta')d\theta'+pF(p)-c_t=\int_{p}^\infty \theta'f(\theta')d\theta'-(1-F(p))p
$$

This simplifies to:

$$
p=c_t
$$

So we have arrived very scientifically at a recurring irrelevance theme: individual productivity does not matter in the make-or-not decision in these models. All that matters is the ultimate price. 

Anyways, what about the market? 

In this case, since the costs of asset creation are $c+e_c$, and everyone for whom $c+e_c\leq p$ creates an asset, we have:
$e_c \leq p - c$ as the cutoff condition, so the volume of assets created is:

$$
G(p-c)
$$

So $1-G(p-c)$ do not create assets. Of those that create assets, a fraction $F(p)$ wind up with productivities less than market price, and a fraction $1-F(p)$ wind up with market productivities greater than price. Hence, Demand-equals-Supply is of the form:

$$
G(p-c)F(p) = (1-G(p-c))(1-F(p))
$$

Or:

$$
1-G(p-c) = F(p)
$$

Note that this is a classic Demand-Suppy model, really. The right-hand-side is increasing in $p$, while the left hand side is decreasing in $p$. 

### Externalities

How would we include externalities in this model? One possibility - common in models of enclosure - is that the costs of creating an asset depend upon aggregate asset creation. This leads to a recursive formulation like:

$$
c = h(G(p-c))
$$

Which should ultimately lead to a function $c=c(p)$. A general procedure might be to first invert, and then see if we can get a differential equation. So, let's try out $h=\rho x$ just to keep it simple. 

$$
c=\rho G(p-c)
$$

This is not in general all that easy to solve. If we suppose that idiosyncratic costs follow a pareto distribution with parameter $\alpha=1$, with $x\geq 1$ we get:

$$
c=\rho \left(1 - \frac{1}{p-c}\right)
$$

The above equation typically has multiple solutions, but seems to be increasing if one focuses on the function with posit

$$
c = \frac{1}{2}(p+\rho+\sqrt{(p-\rho)(p+\rho)+4\rho})
$$
